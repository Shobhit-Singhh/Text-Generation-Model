```python
tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in nltk.sent_tokenize(corpus.lower())]

# Train Word2Vec model
word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=10, min_count=1, workers=4)

# Function to embed a sentence
def embed_sentence(sentence, model):
    return [model.wv[word] for word in nltk.word_tokenize(sentence) if word in model.wv]

# Embed all sentences
embedded_sentences = [embed_sentence(' '.join(sentence), word2vec_model) for sentence in tokenized_sentences]

# Create sequences from pre-embedded sentences
def create_sequences(embedded_sentences, sequence_length):
    X = []
    y = []
    for sentence in embedded_sentences:
        if len(sentence) <= sequence_length:
            continue
        for i in range(len(sentence) - sequence_length):
            sequence = sentence[i:i + sequence_length]
            next_word = sentence[i + sequence_length]
            X.append(sequence)
            y.append(next_word)
    return np.array(X), np.array(y)

sequence_length = 10
X, y = create_sequences(embedded_sentences, sequence_length)

# Debugging: Print the shapes of X and y
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")

X_padded = pad_sequences(X, maxlen=sequence_length, dtype='float32', padding='post')
y_padded = np.array(y)

# Debugging: Print the shapes of X_padded and y_padded
print(f"Shape of X_padded: {X_padded.shape}")
print(f"Shape of y_padded: {y_padded.shape}")

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_padded, test_size=0.2, random_state=42)

# Define model creation functions
def create_simple_rnn_model(input_shape, optimizer):
    model = Sequential()
    model.add(SimpleRNN(100, input_shape=input_shape))
    model.add(Dense(word2vec_model.vector_size, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['accuracy'])
    return model

def create_lstm_model(input_shape, optimizer):
    model = Sequential()
    model.add(LSTM(100, input_shape=input_shape))
    model.add(Dense(word2vec_model.vector_size, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['accuracy'])
    return model

def create_gru_model(input_shape, optimizer):
    model = Sequential()
    model.add(GRU(100, input_shape=input_shape))
    model.add(Dense(word2vec_model.vector_size, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['accuracy'])
    return model

# Function to train and evaluate model
def train_and_evaluate(model, X_train, y_train, X_test, y_test, optimizer_name, model_name):
    history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f'{model_name} with {optimizer_name}: Loss = {loss}, Accuracy = {accuracy}')
    return history

# Train models with different optimizers
optimizers = ['adam', 'sgd', 'rmsprop']
input_shape = (sequence_length, word2vec_model.vector_size)

# Dictionary to store history
histories = {}

for model_name, model_func in [('SimpleRNN', create_simple_rnn_model),
                                ('LSTM', create_lstm_model),
                                ('GRU', create_gru_model)]:
    for optimizer in optimizers:
        print(f"Training {model_name} with {optimizer} optimizer.")
        model = model_func(input_shape, optimizer)
        history = train_and_evaluate(model, X_train, y_train, X_test, y_test, optimizer, model_name)
        histories[f'{model_name}_{optimizer}'] = history

# Plotting
def plot_history(histories, metric='loss'):
    plt.figure(figsize=(12, 8))
    for key in histories:
        plt.plot(histories[key].history[metric], label=key)
    plt.title(f'Model {metric} comparison')
    plt.xlabel('Epochs')
    plt.ylabel(metric.capitalize())
    plt.legend()
    plt.show()

# Plot loss and accuracy
plot_history(histories, 'loss')
plot_history(histories, 'accuracy')
```
